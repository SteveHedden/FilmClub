```python
#This returns the network of a given actor. 
def getNetwork(name):
    node = modernMovies.loc[modernMovies['primaryName'] == name]
    nodeNetwork = pd.merge(node,data, how='left', left_on="tconst",right_on="tconst")
    node = pd.DataFrame(nodeNetwork['primaryName_y'])
    node['freq'] = node.groupby('primaryName_y')['primaryName_y'].transform('count')
    node['Name'] = name
    node = node.drop(node[node['freq'] < 2].index)
    node = node.drop(node[node['primaryName_y'] == name].index)
    return(node)
```

```python
Nick.head()
```

```python
#Get the network of all of Nick's 66 repeat collaborators. 
for x in Nick['primaryName_y']:
    network = getNetwork(x)
    Nick = pd.concat([Nick,network])
    print(x)
```
